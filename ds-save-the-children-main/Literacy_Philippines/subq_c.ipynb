## Q: For attendance as a predictor, what is the minimum level for achieving expected outcome of reading with comprehension? 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('data/merged_Ph.csv', usecols=['ChildrenId', 'Baseline to Endline - Group', 'AttendanceRate', 'ResultBaseline','ResultEndline']).drop_duplicates()
df
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='ResultEndline', y='AttendanceRate')

# Add titles and labels
plt.title('Attendance Rate VS ResultEndline')
plt.xlabel('Baseline to Endline Progress')
plt.ylabel('Attendance Rate')

plt.show()
```

```python
df_filtered = df[df['ResultEndline'] == 'Story with comprehension']

# Plot attendance rates for each baseline-to-endline transition
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_filtered, x='Baseline to Endline - Group', y='AttendanceRate')

# Add titles and labels
plt.title('Attendance Rate Required for Achieving "Reading with Comprehension"')
plt.xlabel('Baseline to Endline Progress')
plt.ylabel('Attendance Rate')

plt.show()
```
